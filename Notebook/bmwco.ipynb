```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('bmw_globalsales_2018_2025.xlsx')
```

```python
df['Avg_Price_EUR'] = df['Avg_Price_EUR'].astype(float)
df['Revenue_EUR'] = df['Revenue_EUR'].astype(float)
df['Date'] = pd.to_datetime(df[['Year','Month']].assign(DAY=1))
```

```python
df['Check_Revenue'] = df['Units_Sold'] * df['Avg_Price_EUR']
```

```python
df.groupby('Date')['Units_Sold'].sum().plot()
```

```python
df.groupby('Region')['Units_Sold'].sum().sort_values(ascending=False)
```

```python
df.groupby('Region')['Units_Sold'].sum().sort_values(ascending=False)
```

```python
df = df.sort_values('Date')
df['Sales_Growth'] = df['Units_Sold'].pct_change()
```

```python
df.groupby('Region')['Avg_Price_EUR'].mean()
```

```python
import matplotlib.pyplot as plt

df.groupby('Region')['Units_Sold'].sum().plot(kind='bar')
plt.show()
```
